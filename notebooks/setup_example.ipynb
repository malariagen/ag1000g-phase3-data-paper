```python
v3_release = release_data()
samplesets_of_interest = ["AG1000G-BF-A", "AG1000G-BF-B"]
```

```python
v3_release.all_wild_sample_sets
```

# 1. Simple example of loading data for a sampleset

```python
gc_mask = v3_release.load_mask("2L", "gamb_colu")
gc_mask
```

```python
pos = v3_release.load_variants("2L", "POS", mask=gc_mask)
pos
```

```python
gt = v3_release.load_sample_set_calldata("2L", samplesets_of_interest, "GT", mask=gc_mask)
```

```python
meta = v3_release.load_sample_set_metadata(samplesets_of_interest, True)
meta.sample_set.value_counts()
```

```python
# or load all samplesets
all_meta = v3_release.load_sample_set_metadata(v3_release.all_sample_sets, True)
all_meta.sample_set.value_counts()
```

# 2. Slightly more complex example of loading data for a species

```python
is_colu = meta.is_coluzzii
is_colu.value_counts()
```

```python
meta_colu = meta.loc[is_colu]
meta_colu
```

```python
gt_colu = gt.compress(is_colu.values, axis=1)
gt_colu

# if not GT use the form:
# gt_colu = da.compress(is_colu.values, gt, axis=1)
```
